```python
import import_ipynb
import wikipedia
import nltk
from numpy import array
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import KFold
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
print("Data is::")
print(wikipedia.summary("Muhammad Ali Jinnah"))
data = (wikipedia.summary("Muhammad Ali Jinnah"))
print(type(data))
def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
             counts[word] += 1
        else:
             counts[word] = 1

    return counts

print( word_count(data))
```

```python
#for finding stop words

stops = set(stopwords.words('english'))
print("Stop Words ::")
print(stops)

#converting data into tokens(seperate words)
word_tokens = word_tokenize(data)


# way 1

filtered_sentence = []

for w in word_tokens:
    if w not in stops:
        filtered_sentence.append(w)
print("\nData Before removing Stop Words::")
print(word_tokens)
print("\nData After removing Stop Words::")
print(filtered_sentence)



# lemitization

lemmatizer = WordNetLemmatizer()
lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_tokens])

print(lemmatized_output)


# K folding for train and test
mydata = array(word_tokens)

# number of k folds
kfold = KFold(n_splits=3, random_state=None)
for train, test in kfold.split(mydata):
    print('\n train: %s, \n \n test: %s' % (mydata[train], mydata[test]))
  
```
